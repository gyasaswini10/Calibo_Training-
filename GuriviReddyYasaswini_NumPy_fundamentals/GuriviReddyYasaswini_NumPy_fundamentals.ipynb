````python
'''
### Task 2.2: Slicing operations (8 points)

Given `arr = np.arange(0, 20)`:

1. Extract elements from index 5 to 15 (inclusive)
2. Extract every 3rd element from the array
3. Reverse the entire array using slicing

Create a 4x4 2D array and:
- Extract the first 2 rows and first 3 columns
- Extract the last 2 rows and last 2 columns
- Extract every other row with all columns
'''
arr = np.arange(0, 20)
print(arr)
print("Extract elements from index 5 to 15 (inclusive)")
print(arr[5:16])
## arr[inclusive:exclusive]
print("Extract every 3rd element from the array")
print(arr[::3])
## from 0 skips 3 indexex
print("Reverse the entire array using slicing")
print(arr[::-1])
## using negative indexing
arr=np.arange(1,17).reshape(4,4)
print(arr)
print("Extract the first 2 rows and first 3 columns")
print(arr[:2,:3])
print("Extract the last 2 rows and last 2 columns")
print(arr[-2:,-2:])
print("Extract every other row with all columns")
print(arr[::2][:2])

````

````python
'''
## Exercise 7: Challenge - Sales Data Analysis (30 points)

You are analyzing sales data for a company with 4 products over 6 months:

```python
sales_data = np.array([
    [120, 135, 150, 145, 160, 175],  # Product A
    [90, 95, 100, 110, 115, 120],    # Product B
    [150, 145, 155, 160, 170, 180],  # Product C
    [80, 85, 82, 88, 92, 95]         # Product D
])
```

**Tasks:**

1. **Calculate total sales for each product** (5 points)
   - Sum across all months for each product

2. **Calculate total sales for each month** (5 points)
   - Sum across all products for each month (Jan to Jun)

3. **Identify the month with the highest total sales** (4 points)
   - Print the month name and total sales

4. **Calculate the growth rate for each product** (6 points)
   - Growth rate = ((Last month - First month) / First month) × 100
   - Print as percentage for each product

5. **Find products with average monthly sales greater than 120k** (4 points)
   - Calculate average for each product
   - Print products meeting the criteria

6. **Calculate quarter-wise sales** (6 points)
   - Q1: January-March
   - Q2: April-June
   - Print sales for each product in each quarter

**Bonus:** Predict next month (July) sales for each product (+5 points)
- Use a simple method: average of last 2 months + average monthly growth

import numpy as np

sales_data = np.array([
    [120, 135, 150, 145, 160, 175],
    [90, 95, 100, 110, 115, 120],
    [150, 145, 155, 160, 170, 180],
    [80, 85, 82, 88, 92, 95]
])
'''
import numpy as np

sales_data = np.array([
    [120, 135, 150, 145, 160, 175],
    [90, 95, 100, 110, 115, 120],
    [150, 145, 155, 160, 170, 180],
    [80, 85, 82, 88, 92, 95]
])
print("Task 1: Total sales for each product")
for i in range(sales_data.shape[0]):
    print("Product",chr(65+i),":",sales_data[i].sum())
print("Task 2: Total sales for each month")
months=["January","February","March","April","May","June"]

for i in range(sales_data.shape[1]):
    print("Month ",(i+1),months[i],":",sales_data[:,i].sum())
print("Task 3: Month with highest total sales")

every_m_totals=sales_data.sum(axis=0)
maxi=every_m_totals.argmax()
print(months[maxi])
print("Month",maxi+1,":",every_m_totals[maxi])
print("Task 4: Growth rate for each product")
#Growth rate = ((Last month - First month) / First month) × 100

for i in range(sales_data.shape[0]):
    growth=((sales_data[i][-1]-sales_data[i][0]) / sales_data[i][0]) * 100
    print("Product",chr(65+i),":",round(growth,2),"%")

print("Task 5: Products with average monthly sales greater than 120k")
for i in range(sales_data.shape[0]):
    avg=sales_data[i].mean()
    if avg>120:
      print("Sales of product",chr(65+i),"with avg",avg)
print("Task 6: Quarter-wise sales")

print("Q1: January–March")
for i in range(sales_data.shape[0]):
    q1=sales_data[i][0]+sales_data[i][1]+sales_data[i][2]
    print("Product",chr(65+i),":",q1)
print("Q2: April–June")
for i in range(sales_data.shape[0]):
    q2=sales_data[i][3]+sales_data[i][4]+sales_data[i][5]
    print("Product",chr(65+i),":",q2)

print("Bonus: Predict next month (July) sales for each product")
print("We use growth rate by comparing first and last month sales and last 2 months average")
for i in range(sales_data.shape[0]):
    last2_avg=(sales_data[i][4]+sales_data[i][5])/2
    growth=(sales_data[i][5]-sales_data[i][0])/sales_data[i][0]
    july=last2_avg+(last2_avg*growth)
    print("Product",chr(65+i),":",round(july,2))
````
